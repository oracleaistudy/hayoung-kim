## 1.1 데이터 사전 작업

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as pit

plt.scatter(bream_length, bream_weight)
plt.xlabel('length') #x축 이름 length로 지정
plt.ylabel('weight') #y축 이름 weight으로 지정
plt.show()
```

```python
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight) #붕어 데이터 산점도
plt.scatter(smelt_length, smelt_weight) #빙어 데이터 산점도
plt.xlabel('length')
plt.ylabel('weight')
plt.show()  
```

```python
#도미와 빙어 데이터를 하나의 데이터로 합침
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
#zip 함수 사용. (리스트에서 원소를 하나씩 꺼내는 함수)
fish_data = [[l, w] for l, w in zip(length, weight)] # length, weight을 zip(하나씩 꺼내서) 각각 l, w에 for문
```

```python
print(fish_data)
```

```python
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

## 1.2. 머신러닝 학습
사이킷 런 패키지 중 k-최근접 이웃 알고리즘을 구현한 클래스인 KNeighborsClassifier 임포트


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier() #KNeighborsClassifier 클래스의 객체 kn 생성
 # 여기에 fish_data와 fish_target을 전달하여 기준을 학습 시킬 것 = Training 훈련
```

```python
kn.fit(fish_data, fish_target) #fit 메서드로 fish_data와 fish_target으로 학습
```

```python
kn.score(fish_data, fish_target) #학습한 모델을 평가하는 메서드 score 호출
```

## 1.3. 학습 시킨 알고리즘 사용해보기

```python
kn.predict([[30, 600   ]]) #도미 데이터로 예측
```

```python
print(kn._fit_X) #학습에 사용된 데이터
```

```python
print(kn._y) #  학습에 사용된 타깃
```
